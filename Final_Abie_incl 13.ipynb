```python
pedestrian = pd.read_csv("C:/Users/abtus/Downloads/count2021-2022.csv")
pedestrian["Date_Time"] = pd.to_datetime(pedestrian["Date_Time"]).dt.date
weekdays = ["Monday", "Tuesday", "Wednesday","Thursday", "Friday"]
sensors = [3,9]

# May 2022
ped_may_2022 = pedestrian[(pedestrian['Year'] == 2022) & (pedestrian['Month'] == 'May') & (pedestrian['Day'].isin(weekdays)) & (pedestrian['Sensor_ID'].isin(sensors)) & (pedestrian['Time']>= 9) & (pedestrian['Time'] <= 17) ] [["Day", "Time", "Date_Time", "Sensor_ID", "Hourly_Counts"]]
ped_may_2022["Date_Time"] =  pd.to_datetime(ped_may_2022["Date_Time"], format="%Y/%m/%d")
ped_may_2022["Date_Time"] = ped_may_2022.Date_Time.dt.strftime('%m-%d')
ped_may_2022 = ped_may_2022.groupby(["Day","Time", "Date_Time", "Sensor_ID", "Hourly_Counts"], as_index=False)['Hourly_Counts'].sum()
ped_may_2022.columns = ["Day","Time", "Date_Time", "Sensor_ID", "Hourly_Counts"]

sensor_3 = ped_may_2022[ped_may_2022['Sensor_ID'] == 3][["Day","Time", "Date_Time", "Hourly_Counts"]]
sensor_3.columns = ["Day","Time", "Date_Time", "Sensor 3 Hourly_Counts"]

sensor_9 = ped_may_2022[ped_may_2022['Sensor_ID'] == 9][["Day","Time", "Date_Time", "Hourly_Counts"]]
sensor_9.columns = ["Day","Time", "Date_Time", "Sensor 9 Hourly_Counts"]

#sensors merged
sensors_merged = pd.merge(sensor_3, sensor_9, how = 'inner', on=["Day","Time", "Date_Time"])

#pearson correlation

pearson_coef = {}
for day_date in sensors_merged['Date_Time'].unique():
    count_sensor3 = sensors_merged[sensors_merged.Date_Time == day_date]['Sensor 3 Hourly_Counts']
    count_sensor9 = sensors_merged[sensors_merged.Date_Time == day_date]['Sensor 9 Hourly_Counts']

    pearson_coef[day_date] = count_sensor3.corr(count_sensor9,method = 'pearson')

# find maximum
max_day_date = max(pearson_coef, key=pearson_coef.get)
max_change = pearson_coef[max_day_date]
min_day_date = min(pearson_coef, key=pearson_coef.get)
min_change = pearson_coef[min_day_date]

max_day = sensors_merged.loc[sensors_merged['Date_Time'] == max_day_date, ['Day']]
max_day = np.unique(max_day)
min_day = sensors_merged.loc[sensors_merged['Date_Time'] == min_day_date, ['Day']]
min_day = np.unique(min_day)

print("Day with the most change in May 2022 is on " + str(max_day_date) + ", a " + str(max_day)  +  ", and the greatest change is " + str((max_change)) + ".")
print("Day with the least change in May 2022 is on " + str(min_day_date) + ", a " + str(min_day) + ", and the least change is " + str((min_change)) + ".")
```
